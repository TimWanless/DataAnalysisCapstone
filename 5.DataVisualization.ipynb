## Demo: How to run an sql query

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here

QUERY = """
SELECT *
FROM master
"""

df = pd.read_sql_query(QUERY,conn)

import matplotlib.pyplot as plt
plt.hist(df['ConvertedComp'])
plt.title('Histogram of Converted Compensation')
plt.ylabel('Count')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
df['Age'].plot(kind='box')
plt.title('Boxplot of Age')
plt.ylabel('Years')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
plt.figure(figsize=(10,3))
plt.scatter(df['Age'],df['WorkWeekHrs'])
plt.title('Scatterplot of Hours Worked by Age')
plt.xlabel('Age')
plt.ylabel('Hours Worked')

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.scatter(df['WorkWeekHrs'],df['CodeRevHrs'],s=df['Age'],alpha=0.3)
plt.title('Bubbleplot of Work Week Hours by Code Review Hours by Age')
plt.xlabel('Code Review Hours')
plt.ylabel('Work Week Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df_next = pd.read_sql_query(QUERY,conn)
df_count=df_next['DatabaseDesireNextYear'].value_counts().rename_axis('DatabaseName').reset_index(name='Count')
df_5=df_count[:5].copy()

plt.pie(df_5['Count'],labels=df_5['DatabaseName'],autopct='%.1f')
plt.title('5 Best Databases to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df_30_35 = pd.read_sql_query(QUERY,conn)

df_30_35_m=df_30_35.groupby('Age').median()

df_30_35_m.plot(kind='bar',stacked=True)
plt.title('Median Work Week Hours and Code Review Hours for 30 to 35 Year Olds')
plt.xlabel('Age')
plt.ylabel('Code Review Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df_45_60 = pd.read_sql_query(QUERY,conn)

df_45_60_m=df_45_60.groupby('Age').median()

df_45_60_m.plot(kind='line')
plt.title('Median Converted Compensation for 45 to 60 Year Olds')
plt.xlabel('Age (Years Old)')
plt.ylabel('Converted Compensation in $')
plt.show()
```
